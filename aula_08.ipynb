##Matplotlib, bliblioteca de analise de dados

```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
```

```python
#oque ja sabemos 
empresas=['Apple','Google','Amazon',
          'Microsoft','Tencent','Visa',
          'McDonalds','Facebook','Alibaba','Louis Vuitton']
```

```python
bilhoes=[947.062,819.573,705.646,
         611.460,214.023,196.526,
         191.032,186.421,169.966,
         124.273]
```

```python
#ja e possivel plotar , so com variaveis
plt.plot(empresas,bilhoes)
```

```python
plt.plot(empresas,bilhoes)
plt.title('empresas mais ricas do mundo')
```

```python
plt.plot(empresas,bilhoes)
plt.title('empresas mais ricas do mundo')
plt.xlabel('empresas')
plt.ylabel('bilhoes')
```

```python
#redmencionameto
plt.rcParams["figure.figsize"]=(15,8)

plt.plot(empresas,bilhoes)
plt.title('empresas mais ricas do mundo')
plt.xlabel('empresas')
plt.ylabel('bilhoes')
```

```python
plt.rcParams["figure.figsize"]=(15,8)

plt.plot(empresas,bilhoes)
plt.title('empresas mais ricas do mundo',fontsize=25)
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
```

```python
plt.rcParams["figure.figsize"]=(15,8)

plt.plot(empresas,bilhoes, color='green')
plt.title('Empresas mais ricas do mundo',fontsize=25, color='red')
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'top', fontsize=15)
```

```python
#adicionadno marcador
plt.plot(empresas,bilhoes, color='green',marker='>')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
```

```python
#adicionadnolinha de grade
plt.plot(empresas,bilhoes, color='green',marker='>')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
plt.grid(True)
```

```python
#mudando o tipo de grafico e colocando leganda acima
plt.bar(empresas,bilhoes, color='red')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
plt.grid(True)
```

```python
#grafico simples
plt.bar(empresas,bilhoes)
```

```python
cores= ['black','dimgray','dimgrey',"gray",'grey','darkgray',
        'darkgrey','silver','lightgray','lightgrey']
```

```python
plt.bar(empresas,bilhoes, color= cores , edgecolor='black')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
plt.grid(True)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-white')
plt.bar(empresas,bilhoes, color= cores , edgecolor='black')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('empresas',fontsize=20)
plt.ylabel('bilhoes',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
plt.grid(True)
```

```python
#barh
plt.style.use('seaborn-white')
plt.barh(empresas,bilhoes, color= cores , edgecolor='black')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('Bilhes',fontsize=20)
plt.ylabel('Empresas',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
plt.grid(True)
```

```python
#scatter
plt.style.use('seaborn-dark')
plt.scatter(empresas,bilhoes, color= cores , edgecolor='black')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.xlabel('Bilhes',fontsize=20)
plt.ylabel('Empresas',fontsize=20)
plt.legend(['bilhoes de dolares'], loc = 'best', fontsize=15)
plt.grid(True)
```
