Esse conjunto de dados mostra as 50 contas mais seguidas do tiktok.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets,layers,models
```

###importando o Dataset

```python
contas = pd.read_csv('/content/List of most-followed TikTok accounts (1).csv')
```

###Selecionando apenas duas colunas para fazer uma comparação de grafico

```python
# Selecionando Colunas Categóricas
# Não usei o Rank
grafico1 = contas.copy()
grafico1 = grafico1[['Rank','Followers\n(millions)']]
grafico1.describe().transpose()
```

```python
(grafico1['Followers\n(millions)'].value_counts() / grafico1.shape[0]) * 100
```

```python
# Visualizando quantidade de seguidores no gráfico
plt.style.use('seaborn-white')
plt.figure(figsize=(15,10))
plt.hist(grafico1['Followers\n(millions)'])
plt.title('Distribuição de quantidades de Seguidores')
plt.xlabel('Seguidores em milhões')
plt.ylabel('Quantidade contas seguidas')
plt.grid(True)
plt.show()
```

###Contas mais seguidas em ordem
